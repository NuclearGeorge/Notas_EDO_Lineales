# $dy/dx=6 xy^2,\;\; y(1)=1/25$

```python
#--------------Importando paquetes------------
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
#---------------------------------------------
# dy/dt como función de t y y respectivamente
def f(y,x):
    dydx = 6*x*y**2
    return dydx

# condición inicial
x0 = 1.0
y0 = 1.0/25.0

# tiempo
x = np.linspace(x0,3)

# Soluci de la EDO
y = odeint(f,y0,x)

# Gráfica de resultado
plt.plot(x,y)
plt.xlabel('x', fontsize=18)
plt.ylabel('y(x)', fontsize=18)
plt.show()
```
