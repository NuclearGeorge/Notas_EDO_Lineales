Por tanto la soluci√≥n se satisface cuando consideramos el valor negativo, es decir

$y(x)=\dfrac{-(x^2+1)\pm \sqrt{x^4+12x^3+2x^2+25}}{2}$

```python
def D(x):
    y = x**4 + 12*x**3 + 2*x**2 + 25
    return y
# tiempo
x = np.linspace(-12,5)
# Gr√°fica de resultado

plt.plot(x,D(x), color='red')
plt.xlabel('x', fontsize=18)
plt.ylabel('D(x)', fontsize=18)
plt.show()
```

```python
#--------------Importando paquetes------------
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
#---------------------------------------------
# dy/dx como funci√≥n de x la variable independiente y y la variable dependiente
def f(y,x):
    dydx = -(2*x*y-9*x**2)/(2*y+x**2+1)
    return dydx

# condici√≥n inicial
x0 = 0
y0 = -3

# tiempo
x = np.linspace(x0,5)

# Soluci de la EDO
y = odeint(f,y0,x)

# Gr√°fica de resultado
plt.plot(x,y, color='blue')
plt.xlabel('x', fontsize=18)
plt.ylabel('y(x)', fontsize=18)
plt.show()
```
